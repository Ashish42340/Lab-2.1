### **Lab Activity 2.1: Exploratory Data Analysis (EDA) and Statistical Visualization of Student Performance Data**

#### **Objective**

- To understand the structure of student performance data using Exploratory Data Analysis (EDA).

- To analyze relationships between different variables such as study time, free time, and final grades (G3).

- To identify patterns, trends, and outliers in the dataset.

- To use statistical visualization techniques (histogram, boxplot, scatter plot) for better interpretation of data.

#### **Requirements**

**Software / Tools**
- Python (Anaconda / Jupyter Notebook / Google Colab)

- Libraries: Pandas, Matplotlib, Seaborn

**Dataset**
- student_sample_20.csv containing student information and performance data.
- Matplotlib for plotting graphs
- Seaborn for statistical visualizations 

```python
import pandas as pd
```

#### **Step 1: Data Collection and Loading**

```python

df = pd.read_csv('student_sample_20.csv', sep=';')
print(df.head())
```

#### **Step 2: Data Exploration**

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print('Categorical Features:', categorical_features)
print('Numerical Features:', numerical_features)  
```

#### **Step 3: Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution of student grades (G3 - final grade):

```python
plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=15, kde=True, color='blue')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.title('Distribution of Final Grades')
plt.show()
```

Correlation Heatmap:

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Features")
plt.show()
```

Impact of Study Time on Final Grade:

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['freetime'], y=df['G3'])
plt.xlabel('free time')
plt.ylabel('final Grade')
plt.title('effect of study time on final grade')
plt.show()
```

Parental Education vs. Student Performance:

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Medu'], y=df['G3'])
plt.xlabel("Mother's Education Level")
plt.ylabel('Final Grade')
plt.title("Effect of Mother's Education on Final Grades")
plt.show()
```
